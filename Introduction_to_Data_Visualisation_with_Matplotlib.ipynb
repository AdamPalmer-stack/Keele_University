```python
#Import functions
import matplotlib . pyplot as plt
```

```python
#Set up data
days = [ 'Mon' , 'Tue' , 'Wed' , 'Thu' , 'Fri']
attendance = [30 , 35 , 32 , 40 , 38]
```

```python
#Create line plot
plt . plot ( days , attendance , marker = 'o', color='crimson')
plt . title ( 'Student Attendance')
plt . xlabel ( 'Days')
plt . ylabel ( 'Number of Students' )
plt . grid ( True )
plt . show ()
```

```python
#New Data
subjects = [ 'Math' , 'Physics' , 'CS' , 'English']
marks = [85 , 78 , 92 , 74]
```

```python
#Create bar chart
plt . bar ( subjects , marks, color='xkcd:baby poop green', edgecolor='navy' )
plt . title ( 'Marks by Subject' )
plt . xlabel ( 'Subjects' )
plt . ylabel ( 'Marks' )
plt . show ()
```

```python
#New data
activities = [ 'Study' , 'Sleep' , 'Leisure' , 'Exercise' ]
hours = [8 , 8 , 5 , 3]
```

```python
#Create Pie Chart
explode = (0.1 , 0 , 0 , 0)
my_colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
plt . pie ( hours , labels = activities , autopct = '%1.1f%%', explode = explode, colors = my_colors )
plt . title ( 'Daily Time Distribution' )
plt . show ()
```
